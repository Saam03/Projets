Exercice 1

```python
import pandas as pd

df = pd.read_csv('ventes_commerces.csv')
```

```python
print(f"Nombre de lignes : {len(df)}")
print(f"Nombre de colonnes : {df.shape[1]}")
print(f"\nNoms des colonnes : {list(df.columns)}")
print(f"\nTypes de données :\n{df.dtypes}")
```

Exercice 2

```python
ca_par_categorie = (
    df.groupby("categorie")["montant_total"]
      .sum()
      .reset_index()
)

print(ca_par_categorie)


categorie_max = ca_par_categorie.iloc[
    ca_par_categorie['montant_total'].idxmax()
]

print(categorie_max)

```

Exercice 3

```python
transactions_by_region = df.groupby('region').size()

region_max = transactions_by_region.idxmax()

print(region_max)
```

```python
amount_by_region = df.groupby('region')['montant_total'].sum()

amount_max = amount_by_region.idxmax()

print(amount_max)
```

Exercice 4

```python
df_filtre = df[df["montant_total"] > 150]

moyenne_par_categorie = (
    df_filtre
    .groupby("categorie")["montant_total"]
    .mean()
)

print(moyenne_par_categorie)
```

Exercice 5

```python
tableau_croise = pd.pivot_table(
    df,
    values="montant_total",
    index="categorie",
    columns="region",
    aggfunc="sum",
    fill_value=0
)

print(tableau_croise)
```

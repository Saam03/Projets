```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../../../data/ventes_commerces.csv')
df.describe()
```

```python
plt.hist(df['montant_total'], bins=10, color='blue', edgecolor='black')
plt.title("Histogramme des montants")
plt.xlabel("Montant")
plt.ylabel("Fréquence")
plt.show()
```

Le montant le plus fréquent des produits est celui compris dans l'intervalle [75;150] avec une fréquence avoisinant les 300. Plus le montant est important moins le produit est acheté.

Exercice 2

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(df.groupby("region")["montant_total"].sum().index, df.groupby("region")["montant_total"].sum().values)
plt.title("Chiffre d'affaires par région")
plt.xlabel("Région")
plt.ylabel("Chiffre d'affaires")
plt.show()
```

Le Nord est la région qui produit le plus de CA de loin avec un turnover avoisinant les 35000 euros.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(df["prix_unitaire"], df["quantite"])
plt.title("Prix unitaire vs Quantité")
plt.xlabel("Prix unitaire")
plt.ylabel("Quantité")
plt.tight_layout()
plt.show()
```

Il n’y a pas de relation négative claire entre le prix unitaire et la quantité achetée. Il n'y a donc pas de corrélation entre le prix et la quantité.

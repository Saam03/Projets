Exercice 1

```python
transactions = [
    {'produit': 'Laptop', 'prix': 899.99, 'quantite': 2, 'client_id': 101},
    {'produit': 'Souris', 'prix': 25.50, 'quantite': 5, 'client_id': 102},
    {'produit': 'Clavier', 'prix': 79.99, 'quantite': 3, 'client_id': 101},
    {'produit': 'Écran', 'prix': 299.99, 'quantite': 1, 'client_id': 103},
    {'produit': 'USB', 'prix': 15.00, 'quantite': 10, 'client_id': 102}
]
```

```python
for transaction in transactions : 
    transaction['total']=transaction['prix']*transaction['quantite']
print(transactions)
```

Exercice 2

```python
transaction_max = max(transactions, key=lambda transaction: transaction["total"])

print(transaction_max)
```

Exercice 3

```python
somme_total = sum(transaction['total'] for transaction in transactions)
print(somme_total)
```

Exercice 4

```python
transactions_over_100 = [
    transaction for transaction in transactions
    if transaction["total"] > 100
]

print(transactions_over_100)
```

Exercice 1

```python
sum(ventes_mensuelles)
```

Exercice 2

```python
import numpy as np

moyenne = np.mean(ventes_mensuelles)
print(moyenne)
```

```python
import statistics

ecart_type = statistics.stdev(ventes_mensuelles)
print(ecart_type)
```

Exercice 3

```python
mois_sup_15000 = [
    i + 1 for i, vente in enumerate(ventes_mensuelles)
    if vente > 15000
]

print(mois_sup_15000)
print(len(mois_sup_15000))
```

Exercice 5

```python
def analyse_ventes(ventes):
    return {
        "total": sum(ventes),
        "moyenne": sum(ventes) / len(ventes),
        "minimum": min(ventes),
        "maximum": max(ventes),
        "mois_sup_15000": sum(i + 1 for i, vente in enumerate(ventes_mensuelles))
    }

print(analyse_ventes(ventes_mensuelles))
```

# Tokenisation

## Fil rouge

Vous savez ce qu'est un LLM. Deuxième étape : préparer le texte pour le modèle. Un LLM travaille sur des nombres (tokens), pas sur du texte brut.

## Contexte

Vous disposez d'un texte brut. Un LLM ne peut pas traiter directement des caractères ou des mots. Il faut transformer le texte en une séquence de nombres (tokens) via un vocabulaire partagé.

## Ce qu'on attend de vous

- Comprendre la différence entre token et mot
- Produire un encodage (liste d'entiers) à partir d'un texte
- Décoder des tokens en texte

## Comment faire

1. Chargez le fichier de données `../../data/sample_text.txt`
2. Créez un vocabulaire simple (par caractère ou par mot)
3. Encodez une phrase en liste d'entiers
4. Décodez la liste d'entiers en texte

```python
# Tokenisation par caractère (simple pour débuter)

# A COMPLETER : Chargez le fichier sample_text.txt
with open('sample_text.txt', 'r', encoding='utf-8') as f:
    texte = f.read()

# A COMPLETER : Créez un vocabulaire (caractère -> id) et (id -> caractère)
caracteres_uniques = sorted(set(texte))
char_to_id = {c: i for i, c in enumerate(caracteres_uniques)}
id_to_char = {i: c for c, i in char_to_id.items()}

print("Taille du vocabulaire :", len(char_to_id))

# A COMPLETER : Encodez une phrase en liste d'entiers
phrase = "Hello"  # Remplacez par une phrase du corpus
tokens = [char_to_id.get(c, 0) for c in phrase]
print("Phrase :", phrase)
print("Tokens :", tokens)

# A COMPLETER : Décodez les tokens en texte
texte_decode = ''.join([id_to_char.get(t, '?') for t in tokens])
print("Texte décodé :", texte_decode)
```

## Contexte

Un LLM (Large Language Model) est un modèle qui prédit le prochain élément d'une séquence de texte. Contrairement à un moteur de recherche qui retrouve des documents existants, un LLM génère du texte nouveau en s'appuyant sur ce qu'il a appris.

## Principe fondamental

Un LLM fonctionne par prédiction du token suivant :

1. On lui donne une séquence de tokens (mots ou sous-parties de mots)
2. Il calcule une probabilité pour chaque token possible du vocabulaire
3. On sélectionne un token (par exemple le plus probable ou par tirage aléatoire)
4. On ajoute ce token à la séquence et on recommence (génération autoregressive)

Exemple : si on donne "Le chat mange", le modèle peut prédire "une" avec une probabilité élevée, puis "souris", etc.

## Comment faire

Lisez les explications ci-dessus. Ensuite, exécutez la cellule suivante qui illustre le principe avec un exemple très simple (bigramme : prédiction du prochain caractère).

```python
# Exemple simplifié : prédiction du prochain caractère (bigramme)
# Ce n'est pas un vrai LLM, mais le principe est le même

texte = "bonjour le monde"
caracteres = list(texte)

# A COMPLETER : Créez un dictionnaire qui compte les paires (caractère actuel, caractère suivant)
paires = {}
for i in range(len(caracteres) - 1):
    paire = (caracteres[i], caracteres[i+1])
    paires[paire] = paires.get(paire, 0) + 1

# Affichez les paires pour comprendre la structure
print("Paires (caractère actuel, suivant) :", paires)

# A COMPLETER : Étant donné un caractère, prédisez le suivant (le plus fréquent)
def predire_suivant(car):
    candidats = {suivant: count 
                 for (actuel, suivant), count in paires.items() 
                 if actuel == car}

    return max(candidats, key=candidats.get) if candidats else None

print(predire_suivant("o"))
```

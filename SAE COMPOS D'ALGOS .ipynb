```python
def est_base(car):
    
    """
    la fonction "est_base" prend en paramètre un caractère et retourne True si ce caractère correspond à une base de l'ADN (est un des caractères A, T, G, C), et False sinon.
    """

    if car=="A" or car=="G" or car=="C" or car=="T":
        return True
    else:
        return False

a=est_base("P")
print(a)
```

```python
def arn_to_codons(sequence):
    
    """
    la fonction "arn_to_codons" prend en paramètre une chaîne de caractères correspondant à de l'ARN et découpe cet ARN en codons. La fonction doit retourner un tableau contenant la liste des codons.
    """

    j= 0
    n = 3
    tab=[]
    while j < len(sequence) and n<=len(sequence) :
        lille=""
        while j < n:
            lille +=str(sequence[j])
            j += 1
        n += 3
        tab.append(lille)   
    return tab



d=arn_to_codons("AGUUAACUA")
print(d)
```

```python
def nextIndice(tab,ind,elements):
    
    """
    La fonction "nextIndice" recherche dans le tableau "tab" à partir de l'indice "ind" et retourne l'indice de la première case du tableau "tab" contenant une valeur de "elements".
    """
    
    while ind<len(tab)-1 and tab[ind] not in elements:
        ind+=1
    return ind
    
tab=["bonjour", "hello", "buongiorno", "ciao", "bye"]
elements=["hello", "bye"]

g=nextIndice(tab,2,elements)
print(g)
```

```python
def decoupe_sequence(seq,start,stop):
    
    """ 
    La fonction "decoupe_sequence" doit découper le tableau "seq" en séquences et retourner un tableau contenant les différents morceaux. Un morceau dans le tableau "seq" est une partie non vide de "seq" comprise entre une valeur de "start" et une valeur de "stop".
    """
    
    tab=[]
    i=0
    while i<len(seq):
        i = nextIndice(seq, i, start)+1
        tab1=[]

        while i<nextIndice(seq, i, stop):
            tab1.append(seq[i])
            i+=1
        tab.append(tab1)
    tab.pop()
    return tab

    
        
        

seq=["val1", "début", "val2", "val3", "end", "val4", "fin", "begin", "val5", "fin", "val6"]
start = ["début", "begin"]
stop = ["fin", "end"]

h=decoupe_sequence(seq, start, stop)
print(h)
```
